```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
from sklearn.metrics.pairwise import linear_kernel
```

```python
data =  pd.read_csv("NMF_2",index_col=None, sep="\t")
```

```python
titles =pd.DataFrame(data['movie_title'])
movie = data.drop(['Unnamed: 0','movie_title'], axis=1)
movie.sample(10)
```

```python
cosine_sim = linear_kernel(movie, movie)
```

```python
cosine_sim = pd.DataFrame(cosine_sim)
cosine_sim.sample(10)
```

```python
indices = pd.Series(movie.index, index=titles)
indices.sample(10)
```

```python
indices.index[0]
```

```python
def recommendations(title, cosine_sim=cosine_sim):
    # Get the index of the movie that matches the title
    idx = indices[title]

    # Get the pairwsie similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies
    sim_scores = sim_scores[1:11]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return titles.iloc[movie_indices]
```

```python
recommendations(('Avatar\xa0',))
```

```python
recommendations(('Spectre\xa0',))
```

```python
recommendations(('Titanic\xa0',))
```
