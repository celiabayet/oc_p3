```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
from sklearn import (manifold, datasets, decomposition, ensemble,discriminant_analysis, random_projection)

#Importing data
data =  pd.read_csv("movie",index_col=None, sep="\t")
```

```python
movie = data.drop(['Unnamed: 0'], axis=1)
print(movie.shape)
movie.sample(10)
```

```python
df = pd.get_dummies(movie)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# apply "le.fit_transform"
df = movie.apply(le.fit_transform)
```

```python
model = manifold.TSNE(n_components=2, perplexity=40, n_iter=3000, init='pca')
tsne = model.fit_transform(df)
```

```python
tsne = pd.DataFrame(tsne)
tsne.sample(10)
```

```python
title = pd.DataFrame(movie['movie_title'])
tsne = pd.concat([tsne, title], axis =1)
tsne.sample(10)
```

```python
tsne.to_csv('tsne', sep='\t', encoding='utf-8')
```
